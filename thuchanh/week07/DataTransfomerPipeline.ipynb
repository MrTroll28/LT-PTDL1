```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
iris = pd.read_csv('iris.csv')
```

```python
iris.drop(columns="Id", inplace=True)
print(iris.head())
```

```python
def my_CubeFunc(x:pd.DataFrame) -> pd.DataFrame:
    return x ** 3
```

```python
x1 = 5
y1 = 7

x2 = -5
y2 = -7

d = (x1-x2)**2 + (y1-y2)**2
print(np.sqrt(d))

dsquare = (x1**2-x2**2)**2 + (y1**2-y2**2)**2
print(np.sqrt(dsquare))

dcube = (x1**3-x2**3)**2 + (y1**3-y2**3)**2
print(np.sqrt(dcube))
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
iris['Species'] = le.fit_transform(iris['Species'])
iris['Species'].unique()
```

```python
test_trans = iris[["SepalLengthCm", "SepalWidthCm", "PetalWidthCm"]]

test_trans = my_CubeFunc(test_trans)

print("DATA FRAME DATA \n")
print('Dataframe \n', iris.head(5))
print('Dataframe after transforming \n', test_trans.head(5))
print('\n')

print("DESCRIBE DATA \n")
print('Dataframe describe \n', iris.describe())
print('Dataframe after transforming \n', test_trans.describe())
print('\n')

print("SKEWS REPORT \n")
print('Dataframe skew \n', iris.skew())
print("testdf skew \n", test_trans.skew())
print('\n')

print("KURTOSIS REPORT \n")
print('Dataframe kurtosis \n', iris.kurtosis())
print("testdf kurtosis \n", test_trans.kurtosis())
print('\n')

print("CORRELATION REPORT \n")
print('Dataframe correlation \n', iris.corr())
print('testdf corr \n', test_trans.corr())
print('\n')
```

```python
from sklearn.preprocessing import MinMaxScaler

test_trans = iris[["SepalLengthCm", "SepalWidthCm", "PetalLengthCm", "PetalWidthCm"]]

scaler = MinMaxScaler()
model = scaler.fit(test_trans)
scaled_data = model.transform(test_trans)

test_trans = pd.DataFrame(data=scaled_data, columns=test_trans.columns)

print("DATA FRAME DATA \n")
print('Dataframe \n', iris.head(5))
print('Dataframe after transforming \n', test_trans.head(5))
print('\n')

print("DESCRIBE DATA \n")
print('Dataframe describe \n', iris.describe())
print('Dataframe after transforming \n', test_trans.describe())
print('\n')

print("SKEWS REPORT \n")
print('Dataframe skew \n', iris.skew())
print("testdf skew \n", test_trans.skew())
print('\n')

print("KURTOSIS REPORT \n")
print('Dataframe kurtosis \n', iris.kurtosis())
print("testdf kurtosis \n", test_trans.kurtosis())
print('\n')

print("CORRELATION REPORT \n")
print('Dataframe correlation \n', iris.corr())
print('testdf corr \n', test_trans.corr())
print('\n')
```

```python
cut_iris = iris[["SepalLengthCm", "SepalWidthCm", "PetalLengthCm", "PetalWidthCm"]]

# 1st Layer transformation
scaled_iris = my_CubeFunc(cut_iris)

print(scaled_iris)

# 2nd Layer transformation
scaler = MinMaxScaler()
model = scaler.fit(scaled_iris)
scaled_data = model.transform(scaled_iris)

# output the transformed data
scaled_iris = pd.DataFrame(data=scaled_data, columns=cut_iris.columns)
print(scaled_iris)
```

```python
processed_iris = pd.concat([scaled_iris, iris["Species"]], axis=1)
print(processed_iris)
```
