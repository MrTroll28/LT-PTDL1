```python
X=iris.iloc[:, 0:4].values
y=iris.iloc[:, 4].values
```

```python
# Metrics
from sklearn.metrics import accuracy_score, precision_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import recall_score, f1_score

from sklearn.model_selection import KFold, train_test_split

from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
gaussian = GaussianNB()
gaussian.fit(X_train, y_train)

Y_pred = gaussian.predict(X_test)
accuracy_nb = round(accuracy_score(y_test, Y_pred) * 100, 2)

acc_gaussian = round(accuracy_score(y_test, Y_pred) * 100, 2)

cm = confusion_matrix(y_test, Y_pred)

accuracy = accuracy_score(y_test, Y_pred)
precision = precision_score(y_test, Y_pred, average='micro')
recall = recall_score(y_test, Y_pred, average='micro')
f1 = f1_score(y_test, Y_pred, average='micro')

print("Confusion Matrix for naive bayes: \n", cm)
print('Accuracy of naive bayes: %.3f' % accuracy_nb)
print('Precision of naive bayes: %.3f' % precision)
print('Recall of naive bayes: %.3f' % recall)
print('F1 Score of naive bayes: %.3f' % f1)
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)

Y_pred = knn.predict(X_test)
accuracy_knn = round(accuracy_score(y_test, Y_pred) * 100, 2)
acc_knn = round(accuracy_score(y_test, Y_pred) * 100, 2)

cm = confusion_matrix(y_test, Y_pred)
accuracy = accuracy_score(y_test, Y_pred)
precision = precision_score(y_test, Y_pred, average='micro')
recall = recall_score(y_test, Y_pred, average='micro')
f1 = f1_score(y_test, Y_pred, average='micro')

print("Confusion Matrix for knn: \n", cm)
print('Accuracy of knn: %.3f' % accuracy_knn)
print('Precision of knn: %.3f' % precision)
print('Recall of knn: %.3f' % recall)
print('F1 Score of knn: %.3f' % f1)
```

```python
plt.figure(figsize=(10, 6))
a_index = list(range(1, 50))
accuracy_scores = []  # Use a list to collect accuracy scores
x = range(1, 50)
for i in a_index:
    model = KNeighborsClassifier(n_neighbors=i)
    model.fit(X_train, y_train)
    Y_pred = model.predict(X_test)
    accuracy_scores.append(accuracy_score(y_test, Y_pred))  # Append accuracy score to the list
a = pd.Series(accuracy_scores)  # Convert the list to a pandas Series
plt.plot(a_index, a, marker='*')
plt.xticks(x)
plt.show()
```

```python
decision_tree = DecisionTreeClassifier()
decision_tree.fit(X_train, y_train)

Y_pred = decision_tree.predict(X_test)
accuracy_dt = round(accuracy_score(y_test, Y_pred) * 100, 2)
acc_dt = round(accuracy_score(y_test, Y_pred) * 100, 2)

cm = confusion_matrix(y_test, Y_pred)
accuracy = accuracy_score(y_test, Y_pred)
precision = precision_score(y_test, Y_pred, average='micro')
recall = recall_score(y_test, Y_pred, average='micro')
f1 = f1_score(y_test, Y_pred, average='micro')

print("Confusion Matrix for decision tree: \n", cm)
print('Accuracy of decision tree: %.3f' % accuracy_dt)
print('Precision of decision tree: %.3f' % precision)
print('Recall of decision tree: %.3f' % recall)
print('F1 Score of decision tree: %.3f' % f1)
```

```python
from sklearn.tree import plot_tree
plt.figure(figsize=(20, 10))
plt.title("Decision Tree")
plot_tree(decision_tree.fit(X_train, y_train), filled=True)
plt.show()
```

```python
results = pd.DataFrame({
    'Model': ['KNN', 'Decision Tree', 'Naive Bayes'],
    'Score' : [acc_knn, acc_dt, acc_gaussian],
    'Accuracy': [acc_knn, acc_dt, acc_gaussian]
})

results_df = results.sort_values(by='Accuracy', ascending=False)
results_df = results_df.reset_index(drop=True)

print(results_df)
```
