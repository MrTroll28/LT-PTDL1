```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('../data/heart_short.xlsx', sheet_name="data")
df
```

```python
X = df[['t_index','c_index']].values # input
y = df[['target']].values # output|
```

```python
X = X.astype(float)
y = y.astype(float)
```

`Random_state`: Giúp cho PY cắt ngẫu nhiên các dòng theo 1 quy tắc. Giá trị khác nhau cắt khác nhau
<br>
`test_size`: tỉ lệ train/test
<br>
`df.index`: Những chỉ số được đưa vào tập test

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,df.index,
                                                    test_size=0.4, random_state=14)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knnModel = KNeighborsClassifier(n_neighbors=5)

knnModel.fit(X_train, y_train)
```

### Đánh giá mô hình
Là kỹ thuật so sánh giá trị thực tế `actual` với giá trị dự báo `predict`
<br>
* `Actual`: được thu thập từ tự nhiên
* `predict`: sinh ra từ mô hình

```python
y_test_predicted =knnModel.predict(X_test)
y_test_predicted

knnModel.score(X_test, y_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_test_predicted)
```
